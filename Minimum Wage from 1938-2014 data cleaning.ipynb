```python
import requests
r = requests.get('https://bebusinessed.com/history/history-of-minimum-wage/').text #Gets min wage html text from the site
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r, 'html.parser')
paragraphs = soup.find_all('p')
for row in paragraphs:
    print row
    print "\n"
```

```python
min_wage_soup = paragraphs[11:33]
min_wage_list = []
for row in min_wage_soup:
    wage_raise = row.text.encode('ascii')
    min_wage_list.append(wage_raise)
print min_wage_list
```

```python
# lets remove special characters
char_list = [ "(" , ")" , "$" , "/hr", ":"]
min_wage_no_char = []
for row in min_wage_list:
    for char in char_list:
        row = row.replace(char,"")
    min_wage_no_char.append(row)
print min_wage_no_char
```

```python
# Now lets switch the month to dates
months = {"October": '10', "January": '1', "March": '3', "September": '9', 
          "February": '2', "May": '5', "April": '4', "July": '7'}
min_wage_no_month = []
for row in min_wage_no_char:
    for month in months:
        row = row.replace(month, months[month])
    min_wage_no_month.append(row)
print min_wage_no_month
```

```python
# Now lets remove the presidents names
presidents = ["FDR", "Truman", "Eisenhower", "Kennedy", "Johnson", "Nixon", 
              "Ford", "Carter", "Bush", "Clinton", "GW", "Obama"]
wage_raises = []
for row in min_wage_no_month:
    for president in presidents:
        row = row.replace(president, "")
    wage_raises.append(row)
print wage_raises
```

```python
# I will now split the string on the whitespace
wage_data = []
for item in wage_raises:
    split_item = item.split(" ")
    wage_data.append(split_item)

for row in wage_data:
    print row
```

```python
# We are left with a blank space in the 2nd index value, lets get rid of it by changing the value to "1".
# This will be helpful in converting the values to datetime objects.
# We also have an extra blank space in the 2008 and 2007 rows, so lets delete that
for row in wage_data:
    row[2] = "1"
del wage_data[20][3]
del wage_data[19][3]

for row in wage_data:
    print row
```

```python
min_wages = []
for row in wage_data:
    month = row[0]
    year = row[1]
    day = row[2]
    min_wage = float(row[3])
    adj_wage = float(row[4])
    min_wages.append([day +"/" + month + "/" + year, min_wage, adj_wage])
print min_wages
```
