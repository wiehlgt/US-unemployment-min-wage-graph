# US Unemployment and Minimum Wage graphing

Here is a quick project showing minimum wage increase on a plot of US unemployment rate, as well as youth unemployment rate. There is a debate among ecomomists on whether minimum wage increases, decreases, or has no effect on the unemployment rate. Studies on this effect are of great importance, however this is not one of those. Here is a quick look.

```python
import pandas as pd
import matplotlib.pyplot as plt

wage_data = pd.read_csv('wagedata.csv')
youth = pd.read_csv('us_youth_unemployment.csv')
wage_data = wage_data.drop(['Unnamed: 0'], axis=1)
wage_data['date'] = pd.to_datetime(wage_data['date'])
wage_data
```

```python
youth['DATE'] = pd.to_datetime(youth['DATE'])
youth.columns = ['date', 'unemploy']
youth
```

```python
%matplotlib inline
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

start = youth['date'][0]
end = youth['date'][42]
fig, ax1 = plt.subplots()

ax1.plot(wage_data['date'], wage_data['wage_adjusted'],'bo')
ax1.plot(wage_data['date'], wage_data['wage'],'or')
ax1.set_ylabel("Minimum Wage (Dollars)")
ax1.set_xlim([start, end])
ax1.legend(loc='upper center')

ax2 = ax1.twinx()
ax2.plot(youth['date'], youth['unemploy'])
ax2.set_ylabel("Youth Unemployment Rate (%)")

plt.show()
```

It is difficult to discern the effects on youth unemployment simply from this graph. One things that sticks out is how despite the increase of the minimum wage from 1974-1982, adjusted for inflation the minimum wage was decreasing. This really shows the mid to late 70's inflation period.

```python
unemployment = pd.read_csv('us_unemployment_1948_2014.csv')
unemployment = unemployment.drop(['Unnamed: 0'], axis=1)
unemployment['year'] = pd.to_datetime(unemployment['year'])
unemployment
```

```python
%matplotlib inline
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

start = unemployment['year'][0]
end = unemployment['year'][803]
fig, ax1 = plt.subplots()

ax1.plot(wage_data['date'], wage_data['wage_adjusted'],'bo')
ax1.plot(wage_data['date'], wage_data['wage'],'or')
ax1.set_ylabel("Minimum Wage (Dollars)")
ax1.set_xlim(start, end)

ax2 = ax1.twinx()
ax2.plot(unemployment['year'], unemployment['unemployment'])
ax2.set_ylabel("Unemployment Rate (%)")

plt.show()
```

Again, it is difficult to tell what the effects are from raising the minimum wage on unemployment rate, telling us that there are certainly many factors at play.
